### Justification for Dropped Features 

During the feature selection phase of the Airbnb price prediction project, several columns were removed based on quantitative and qualitative assessment. The primary motivations were high missing rates, irrelevance to the predictive objective, redundancy, and excessive preprocessing complexity without proportional benefit. Although, many features were removed for the stated reasons, some major removed features and their justifications are summarized below.

Removed Features & Reason for Removal
* host_about-->Contains long unstructured text with ~46% missing values. Extracting meaningful signal would require complex NLP techniques, which exceed the intended scope and computational benefit for a pricing model.
* neighborhood-->Approximately 61% of values missing. Sparse representation would introduce noise rather than useful signal, and imputation would distort locality information already captured by structured geographic fields.
* neighborhood_highlights-->Free-form descriptive text with no standardized meaning, contributing minimal predictive value and increasing preprocessing complexity.
* host_location-->Contains inconsistent location formats and ~23% missing values. Provides information about where the host lives rather than where the listing is located. Listing-location-based features serve this purpose more effectively.
* host_neighbourhood-->Approximately 18% missing values and redundant relative to precise geographic attributes such as latitude, longitude, zipcode, and neighbourhood_cleansed. Low expected correlation with pricing once more direct location features are included.
General Principles Applied

The decision to remove features adheres to the following guidelines of machine learning feature engineering:

High missingness with low recovery value justifies exclusion.

Unstructured text requiring extensive processing is removed unless central to the model objective.

Features that are weak proxies for stronger existing attributes are considered redundant and removed.

Model interpretability is prioritized, avoiding unnecessary noise.

Conclusion

The removed attributes would contribute limited predictive benefit while significantly increasing preprocessing complexity. Eliminating them results in a cleaner, more reliable feature set aligned with the project objective of accurate pricing prediction and interpretable ML modeling.

### Add another column 'has_reviews' to infer if the said Airbnb has been reviewed and to deal with NaN values.

```python
df1['has_reviews']=df1['review_scores_rating'].notna().astype('int64')
review_cols=[
    'review_scores_rating','review_scores_accuracy','review_scores_cleanliness',
    'review_scores_checkin','review_scores_communication','review_scores_location',
    'review_scores_value','reviews_per_month'
]
df1[review_cols]=df1[review_cols].fillna(0)
```

### Replace the NaN values in 'host_response_time' with "unknown" to signify that there is no definite response time from the host.

```python
df1['host_response_time'].replace(np.nan,"unknown",inplace=True)
df1['host_response_time'].head()
```

### Dealing with missing values of 'hosting_response_rate' & 'host_acceptance_rate'

```python
df1['host_response_rate']=df1['host_response_rate'].str.replace('%','',regex=False)
df1['host_response_rate']=pd.to_numeric(df1['host_response_rate'],errors='coerce')
med1=df1['host_response_rate'].median()
df1['host_response_rate'].replace(np.nan,med1,inplace=True)
```

```python
df1['host_acceptance_rate']=df1['host_acceptance_rate'].str.replace('%','',regex=False)
df1['host_acceptance_rate']=pd.to_numeric(df1['host_acceptance_rate'],errors='coerce')
med2=df1['host_acceptance_rate'].median()
df1['host_acceptance_rate'].replace(np.nan,med2,inplace=True)
```

```python
df1[['host_acceptance_rate','host_response_rate']].isnull().sum()
```

### We move to the Target Variable 'price' and drop the rows where price is not mentioned as that data is considered incomplete.
#### Reason for dropping : Filling with mean or median will only result in unnatural artificial data for target which is not desired.

```python
df1=df1.dropna(subset='price')
df1['price']=df1['price'].replace('[$\,]','',regex=True).astype(float)
df1.isnull().sum()
```

### Dealing with missing 'first_review' & 'last_review'

```python
df1['first_review_missing']=df1['first_review'].isna().astype(int) ##missing indicators / flags
df1['last_review_missing']=df1['last_review'].isna().astype(int)
```

```python
df1['first_review']=pd.to_datetime(df1['first_review'])
df1['last_review']=pd.to_datetime(df1['last_review'])
```

#### Replace missing dates with neutral placeholders.

```python
min_date=df1['first_review'].min()
df1['first_review'] = df1['first_review'].fillna(min_date)
df1['last_review'] = df1['last_review'].fillna(min_date)
```

### Feature engineering

```python
#AI💀😑
df1['days_since_first_review']=(pd.Timestamp.today()-df1['first_review']).dt.days 
df1['days_since_last_review']=(pd.Timestamp.today()-df1['last_review']).dt.days
```

```python
df1['has_availability_miss']=df1['has_availability'].isna().astype(int) #missing indicator/flag
df1['has_availability'].replace(np.nan,'unknown',inplace=True)
df1['has_availability']=df1['has_availability'].map({'t':1,'f':0,'unknown':-1}) ##map the values to integers
```

### Similarly with host_is_superhost we deal with the 3 NaN values.

```python
df1['superhost_missing']=df1['host_is_superhost'].isna().astype(int)
df1['host_is_superhost'].replace(np.nan,'unknown',inplace=True)
df1['host_is_superhost']=df1['host_is_superhost'].map({'t':1,'f':0,'unknown':-1})
```
